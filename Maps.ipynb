```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import seaborn as sns
sns.set_theme(context='paper', style='white', palette='colorblind',font='Helvetica')
```

```python
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='l',
            width=2E6, height=3E6, 
            lat_0=71.7069, lon_0=-42.6043)
#m.etopo(scale=1, alpha=1)
#m.shadedrelief()
m.drawcoastlines(linewidth=.8, linestyle='solid', color='k', antialiased=1, ax=None, zorder=None)
m.drawlsmask(land_color='w', ocean_color='w', resolution='h')
m.drawmapboundary(color='k', linewidth=1.0, fill_color=None, zorder=None, ax=None)
#parallels = np.arange(0.,81,5.)
#m.drawparallels(parallels,labels=[False,True,True,False])
#meridians = np.arange(10.,351.,10.)
#m.drawmeridians(meridians, color='k', textcolor='k', linewidth=1.0,labels=[False,True,True,False])
x1, y1 = m(-42.32, 75.1) # NGRIP
x2, y2 = m(-51.36, 77.27) # NEEM
x3, y3 = m(-38.28, 72.35) # GISP2

# Map (long, lat) to (x, y) for plotting
#x, y = m(-122.3, 47.6)
plt.plot(x1, y1, 'or', markersize=6)
plt.text(x1, y1, ' NGRIP', fontsize=14)
plt.plot(x2, y2, 'or', markersize=6)
plt.text(x2, y2, ' NEEM', fontsize=14)
plt.plot(x3, y3, 'or', markersize=6)
plt.text(x3, y3, ' GISP2', fontsize=14)
plt.savefig('Greenland.jpg', dpi=1000, bbox_inches='tight')

# c (crude), l (low), i (intermediate), h (high), f (full) or None
```
